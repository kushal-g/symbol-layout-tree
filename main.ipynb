```python
import latex2mathml.converter
import xml.etree.ElementTree as ET
```

```python
f = open("./quad_quad.svg.xml")
svg_content = f.read()
f.close()


f = open("./quad_quad.mml")
mml_content = f.read()
f.close()
```

```python
svg_root = ET.fromstring(svg_content)
mml_root = ET.fromstring(mml_content)
```

```python
from typing import List
from xml.etree.ElementTree import Element

def tree_print(root: Element, depth=0, keys:List[str] = []):
    tag = root.tag.replace("{http://www.w3.org/2000/svg}","")
    k = []
    
    for key in root.keys():
        if root.get(key):
            k.append(root.get(key))

    if tag == "g":
        print("\t"*depth, k)

    for child in root:
        tree_print(child, depth+1, keys)

tree_print(svg_root, 0, ["data-mml-node"])
```

```python
tree_print(mml_root, 0)
```
